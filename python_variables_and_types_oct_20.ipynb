````python
current = int(input("Enter current year: "))
birth = int(input("Enter birth year: "))

age = current - birth
future = age + 5

print("Now:", age)
print("In 5 yrs:", future)

# –Ω–∞–ø—Ä–∏–∫–ª–∞–¥

Enter current year: 2025
Enter birth year: 2008
Now: 17
In 5 yrs: 22

# –∞–±–æ

current = 2025
birth = 2008

age = current - birth
future = age + 5

print("Now:", age)
print("In 5 yrs:", future)

Now: 17
In 5 yrs: 22
````
